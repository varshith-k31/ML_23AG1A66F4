```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([1, 8])
y = np.array([3, 10])
plt.plot(x, y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='blue',linestyle=':',marker='s',linewidth=3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("line chart")
plt.legend(['my data series'])
plt.grid(True)
plt.show()
```
